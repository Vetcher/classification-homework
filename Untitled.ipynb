```python
def foldl(f, x0, lst):
    if not lst:
        return x0
    return foldl(f, f(x0, lst[0]), lst[1:])

def foldr(f, x0, lst):
    if not lst:
        return x0
    return f(lst[0], foldr(f, x0, lst[1:]))
```

```python
def foldl2(f, x0, lst):
    return foldr(lambda x, y: lambda t: y(f(t, x)), lambda x: x, lst)(x0)

def foldr2(f, x0, lst):
    return foldl(lambda x, y: lambda t: f(y, x(t)), lambda x: x, lst)(x0)

def foldl3(f, x0, lst):
    return foldr(lambda x,g: (lambda z: g(z,x)), f, lst)(x0)

f = lambda x, y: '(%s, %s)' % (x, y)
foldl3(f, 0, [1, 2, 3])
```

```python
def our_vectorize(f):
    def new_f(matr):
        result = np.zeros_like(matr)
        for y in range(matr.shape[0]):
            for x in range(matr.shape[1]):
                result[y, x] = f(matr[y, x])
        return result
    
    return new_f

@our_vectorize
def foo(x):
    if x > 5:
        return 1
    return 0

```

```python
def foo(k):
    res = 1
    for i in range(k):
        res *= (i + 1)
    return res

pool = Pool(2)
```

```python
def fact_seq():
    for r in map(foo, [100000, 200000, 3000, 1000, 10000]):
        pass
        
%time fact_seq()
```

```python
def fact_par():
    for r in pool.imap(foo, [100000, 200000, 3000, 1000, 10000]):
        pass
        
%time fact_par()
```

```python
for r in pool.imap(foo, [100, 200, 300, 100]):
    print(r)
```
